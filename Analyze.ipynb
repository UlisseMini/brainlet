```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import os.path
import glob
```

```python
tasks = os.listdir('data')
dataframes = [
    pd.concat([pd.read_csv(file) for file in glob.glob(os.path.join('data', task) + '/*.csv')])
    for task in tasks
]
```

```python
def analyze(df):
    right = df.query("answer == correct_answer")
    wrong = df.query("answer != correct_answer")
    n_correct = len(right)
    n_wrong = len(wrong)
    print(f'got {n_correct} right and {n_wrong} wrong ({n_correct/len(df)*100:.1f}% accuracy)')
    
    _ = plt.plot(df.time)
    plt.show()
```

```python
for task, df in zip(tasks, dataframes):
    print(task)
    print(f'average time/question: {df.time.mean():.2f}s')
    analyze(df)
```
